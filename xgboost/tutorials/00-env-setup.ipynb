## Environment Setup for XGBOOST

In this tutorial, we will guide you through the steps of setting up your environment on major platforms (Mac OS, Linux, and Windows). We will cover
    - The installation of the **Miniforge** package distribution
    - Creating a virtual environment
    - Installing the **XGBoost** library.

We will begin by installing the package distribution on Mac OS, Linux, and Windows. Then, we will create a virtual environment using the `conda` (or `mamba`) package manager—steps that are consistent across all platforms—and proceed to install **XGBoost** using both `mamba` and `pip`. Additionally, we will cover the installation of the Jupyter Lab IDE (optional). Finally, we will test our environment and the installed libraries to ensure everything is functioning correctly.

### Installing Miniforge on macOS

1. **Install Homebrew**: If you haven't installed Homebrew yet, you can do so by running the following command in your terminal:

   ```sh
   /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
   ```

**After installation**: Optionally you can run the following two commands to make ensure that Homebrew installation is working correctly and up to date:
   ```sh
   brew doctor
   brew update
   ```

2. **Install Miniforge**:  You can install Miniforge using Homebrew with the following command:

   ```sh
   brew install --cask miniforge
   ```

**Alternative: Using the Graphical Installer**

If you prefer a more user-friendly approach, you can download the graphical installer for Miniforge directly from the [official Miniforge download page](https://conda-forge.org/download/).

After downloading the installer open the `.pkg` file and follow the on-screen instructions to complete the installation.


### Installing Miniforge on Linux

1. **Install Dependencies**: Before installing Miniforge, ensure your system has the necessary dependencies installed by running the next commands:
  - **For Debian distributions**
     ```sh
     sudo apt-get update
     sudo apt-get install -y curl bzip2
     ```
  - **For Fedora-based systems**
     ```sh
     sudo dnf update
     sudo dnf install -y curl bzip2
     ```

2.	**Install Miniforge**: Miniforge is a minimal conda installer that is compatible with Linux distributions.
  - Download and install Miniforge:
     ```sh
     curl -L -O https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-Linux-x86_64.sh
     bash Miniforge3-Linux-x86_64.sh
     ```

3.	Using the Graphical Installer: If you prefer to use the graphical installer of Miniforge, you can download it from the [official Miniforge download page](https://conda-forge.org/download/).

After downloading, make the installer executable and run it:
   ```sh
   chmod +x Miniforge3-Linux-x86_64.sh
   ./Miniforge3-Linux-x86_64.sh
   ```
### Installing Miniforge on Windows

1. **Install Dependencies**: Before installing Miniforge, ensure that your system has the necessary package managers installed. There are three popular package managers on Windows:
   - `winget`: Available for Windows 11 and later
   - `choco`
   -  `scoop`. You can use any of these to install Miniforge.

   **Using `winget`**:
   - If you don't have `winget` installed, you can install it through the Microsoft Store or from [GitHub](https://github.com/microsoft/winget-cli/releases).
   - To install Miniforge using `winget`, run the following command:
     ```sh
     winget install --id=conda-forge.miniforge
     ```

   **Using `choco` (Chocolatey)**:
   - If you don't have Chocolatey installed, you can install it by running the following command in PowerShell as Administrator:
     ```sh
     Set-ExecutionPolicy Bypass -Scope Process -Force; [System.Net.ServicePointManager]::SecurityProtocol = [System.Net.ServicePointManager]::SecurityProtocol -bor 3072; iex ((New-Object System.Net.WebClient).DownloadString('https://community.chocolatey.org/install.ps1'))
     ```
   - Once Chocolatey is installed, you can install Miniforge by running:
     ```sh
     choco install miniforge
     ```

   **Using `scoop`**:
   - If you don't have `scoop` installed, you can install it by running the following command in PowerShell:
     ```sh
     Set-ExecutionPolicy RemoteSigned -Scope CurrentUser
     irm get.scoop.sh | iex
     ```
   - After installing Scoop, install Miniforge using:
     ```sh
     scoop install miniforge
     ```

3. **Install Miniforge Using the Graphical Installer**: If you prefer to download and install Miniforge using the graphical installer wihtout a package manager, you can do so as follows:
   - Download Miniforge from the [official Miniforge download page](https://conda-forge.org/miniforge/windows).
   - Once downloaded, run the installer, simply double-click on the downloaded `.exe` file  and follow the on-screen instructions to complete the installation.

After completing the installation, you can start using Miniforge for managing your Python environments on Windows.

## Creating Virtual Development Environment

### Using Conda or Mamba

1. **Create Virtual Environment**
   ```sh
   # Create a separate environment
   conda create -n xgb-env          # choose any name you prefer
   conda activate xgb-env
   ```

2. Install XGBoost Using `Conda` | `Mamba`:

   - **Configure `Conda`: If you are using `conda` mostly, preferably configure it to use `conda-forge` channel to install packages from like this:
   
   ```sh
   # Add conda-forge channel
   conda config --env --add channels conda-forge
   ```

   You may check the configuration using this command:

   ```sh
   conda config --show
   ```
   
  - **Mamba preconfigured with `conda-forge`**: If you are using `mamba` skip the previous step, since `mamba` comes preconfigured with `conda-forge` channel.

3. **Install XGboost**: Now you are ready to use either `conda` or `mamba`:
   - **General Installation**: You can proceed with the installation like this:
     ```sh
     mamba install py-xgboost  --yes
     ```
   - **CPU** only:
     ```sh
     mamba install py-xgboost-cpu --yes
     ```
   - **GPU** Support with NVIDIA: **Not available on Windows** 
     ```sh
     mamba install py-xgboost-gpu  --yes
     ```

4. **Check the Installation**: You can check whether XGBoost installed correctly and what version is installed; to do that run this:

   ```sh
   python -c "import xgboost; print(xgboost.__version__)"
   ```

   Or list all the installed packages in the active environment:

   ```sh
   mamba list
   ```

This command will list all the installed packages and the dependencies. Or may just want to check the version of the installed polars like this: 

   ```sh
   mamba list | grep -E 'xgboost'
   ```

Or on windows:
   ```sh
   mamba list | findstr "xgboost"
   ```
You may also wish to check other information, to do that run the next command:

   ```sh
   mamba info
   ```

## Installing Using `pip`

> **Note**

1. **Installing XGBoost** with `pip`: This tools gives the option to install the full or minimal XGBoost installation:

## Optional: Generating the Environment Requirements

If plan to reproduce this environment, you can do that by generating `requirements.txt` file using pip as follows:

   ```sh
   pip list --format=freeze > requirements.txt
   ```

To recreate the environment use this command:

   ```sh
   pip install -r requirements.txt
   ```

**Or you can do that using conda**

   ```sh
   conda env export > environment.yml
   ```

Later you can recreate the environment with the same dependencies like this:

   ```sh
   conda env create -f environment.yml
   ```

**Combining Conda and Pip**

If your environment uses both Conda and pip packages, you can do the following:

1. Export Conda Packages:
   ```sh
   conda list --export > conda_requirements.txt
   ```

2. Recreate the environment like this:

   ```sh
   conda create --name myenv --file conda_requirements.txt
   ```
